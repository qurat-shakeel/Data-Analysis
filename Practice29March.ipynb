```python
import pandas as pd
data = pd.read_csv('train.csv')
data
```

```python
# no. of female passengers
female_passengers = data[data['Sex']== 'female']
print(len(female_passengers))
```

```python
# number of females survived
female_survived = female_passengers[female_passengers['Survived']==1].shape[0]
print("Number of females survived :",female_survived)
```

```python
# no. of male passengers
male_passengers = data[data['Sex']== 'male']
print(len(male_passengers))
```

```python
# number of males survived
male_survived = male_passengers[male_passengers['Survived']==1].shape[0]
print("Number of males survived :",male_survived)
```

```python
# bar graph
import matplotlib.pyplot as plt

survivors_by_gender = data.groupby(['Sex','Survived']).size().unstack()
survivors_by_gender.plot(kind='bar',stacked=True)

plt.title('Distribution of survivors by gender')
plt.xlabel('Gender')
plt.ylabel('Number of passengers')
plt.xticks(rotation = 0)
plt.legend(['Did not Survived','Survived'])

plt.show()
```

```python
import seaborn as sns
survivors_by_gender = data.groupby(['Sex','Survived']).size().unstack()
sns.heatmap(survivors_by_gender, annot=True, fmt = 'd',cmap='YlGnBu')
plt.title('Distribution of Survivors by gender')
plt.xlabel('Survived')
plt.ylabel('gender')
plt.show()
```
